## Análise de interferência da renda na utilização da internet 

___
## Introdução

Análise da interferência da renda no acesso e utilização dos serviços de internet por meio de dados referentes ao acesso à Internet realizado dos seguintes modos: internet em domicio, internet via 3G,  através de conexão discada por linha telefônica e internet via banda larga.

Pergunta: como a renda interfere no acesso e utilização dos serviços de internet pelo território nacional ?

## Conclusão 

   Ao analisar como o acesso e utilização dos serviços de internet se comporta para as diferentes classes sociais, percebe-
se que não é possível generalizar entre os diversos tipos de conexão ao serviço e portanto parte a necessidade que decompo-lo em alguns tipos, no caso foram nas formas de pessoas que utilizam os serviços de: Internet em domicilio, 3G, Banda larga e Internet via conexão discada por linha telefonica. Sendo o último descartado por ter índices muitos baixos e não favorecer a pesquisa
    
   ### Internet em domicílio
    
   Analizando o uso de internet em domicilio percebe-se que há uma concentração nas classes mais altas, portanto pode ser considerado um serviço caro. Apesar de serem pequenas as porcentagens para as classes baixas (abaixo de 30% para a classe vunerável por exemplo) observando em relação a renda dessas classes e observando também que a quase totalidade das classes altas a possuem pode-se dizer que é considerado um serviço essencial. Ao análisar com o passar dos anos, percebe-se que em 2014 houve um aumento de 7,82% no uso do serviço, concentrado em grande parte nas classes baixas, logo o serviço barateou. Já em 2015 não houve grandes mudaças. 
   Por ter não ter ficado acessível a todos, pode-se dizer que o serviço de internet em domicilio sofre interferência com a variação da renda, caracterizando-se como um serviço para rendas mais altas.
   
   ### Internet 3G 
    
   Já para o serviço de 3G, ao percebe-se que em 2013 o serviço não era muito utilizado por nenhuma das classes, porém também pode ser considerado caro pois concentra-se nas classes altas. De 2013 para 2014 o serviço cresceu muito (18,87%), de tal modo que as classes baixas atingem as classes altas, portanto barateou a ponto de ficar acessível a todos, o serviço é considerado essencial. Em 2015 não houve grandes mudanças. 
   Por ter ficado acessível a todos, pode-se dizer que o serviço de 3G não sofre interferência com a variação da renda, a partir do ano 2014.
    
   ### Internet Banda Larga
    
   Para o serviço de banda larga, nota-se que grande parte da população utiliza-a, portanto não é considerada um serviço caro, mas não barato o suficiente para atingir todas as renda igualmente. Em 2014 houve uma queda de 4,94%, localizada mais nas classes baixas, portanto o serviço sofreu um pequeno aumento em seu preço. Para 2015 não houve grandes mudanças.
   Por ter não ter ficado acessível a todos, pode-se dizer que o serviço de Banda Larga sofre interferência com a variação da renda, caracterizando-se como um serviço para rendas mais altas.
